```python
from EvenTem import Ricom
```

```python
filename = 'NU-1000.electron'
DetectorSize = 108
repetitions = 1
N = 108
```

```python
ricom = Ricom(nx=N,ny=N,repetitions=repetitions,filename=filename)
ricom.DetectorSize = DetectorSize
KS = 5
rotation = 90
ricom.SetKernel(kernelsize=KS,rotation=rotation)
ricom.Run()
```

```python
ricom.PlotImage()
```
