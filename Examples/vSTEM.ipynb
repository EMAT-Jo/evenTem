```python
from EvenTem import Pacbed,vSTEM,Var
```

```python
filename = 'NU-1000.electron'
DetectorSize = 108
repetitions = 1
N = 108
```

```python
pb = Pacbed(nx=N,ny=N,repetitions=repetitions,filename=filename)
pb.DetectorSize = DetectorSize
# pb.DwellTime = dt #not relevant for simulated data
pb.Run()
```

```python
pb.PlotImage(log=True)
```

```python
vs = vSTEM(nx=N,ny=N,repetitions=repetitions,filename=filename)
vs.DetectorSize = DetectorSize
vs.InnerRadia = [108//4]
vs.OuterRadia = [108]
# vs.Centers = [pb.GetCOM()]

vs.PlotDetector(pacbed=pb)
```

```python
vs.Run()
```

```python
vs.PlotImage()
image_npy = vs.image # access the image data as a numpy array
```

```python
var = Var(nx=N,ny=N,repetitions=repetitions,filename=filename)
var.DetectorSize = DetectorSize
# var.Centers = pb.GetCOM()
var.Run()
var.PlotImage()
```
